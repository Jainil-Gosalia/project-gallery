```python
import numpy as np

def positional_encoding(max_seq_len: int, d_model: int) -> np.ndarray:
    """
    Generates sinusoidal positional encoding matrix.

    Args:
        max_seq_len (int): Max length of input sequences.
        d_model (int): Embedding dimension size.

    Returns:
        np.ndarray: Positional encoding matrix of shape (max_seq_len, d_model)
    """
    PE = np.zeros((max_seq_len, d_model))
    position = np.arange(max_seq_len)[:, np.newaxis]                   # (max_seq_len, 1)
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))  # (d_model // 2,)

    PE[:, 0::2] = np.sin(position * div_term)  # Apply sin to even indices
    PE[:, 1::2] = np.cos(position * div_term)  # Apply cos to odd indices

    return PE
```

```python
import matplotlib.pyplot as plt

# Generate encodings
pe = positional_encoding(max_seq_len=100, d_model=16)

# Plot a few dimensions
plt.figure(figsize=(12, 6))
for i in range(4):
    plt.plot(pe[:, i], label=f"Dimension {i}")
plt.title("Sinusoidal Positional Encoding (First 4 Dimensions)")
plt.xlabel("Position")
plt.ylabel("Encoding Value")
plt.legend()
plt.grid(True)
plt.show()
```
